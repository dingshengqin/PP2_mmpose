## 测试集上评估模型精度

### 指标说明
- class	检测类别，这里是 "person"，表示检测目标为人。
- gts	Ground Truth Samples（真实样本数），在这个例子中为 4021，表示在数据集中实际存在的人。
- dets	Detected Samples（检测到的样本数），在这个例子中为 13730，表示模型检测到的人数。
- recall	召回率，值为 0.761，表示模型检测到的真实目标数占所有真实目标数的比例。计算公式为：
- ap	平均精度（Average Precision），值为 0.612，表示模型在不同召回率下的精度的平均值。AP 是衡量目标检测模型性能的常用指标，值越高表示模型性能越好。
